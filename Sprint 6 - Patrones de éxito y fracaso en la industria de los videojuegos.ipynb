##### Columna `year_of_release`


Para completar los valores faltantes en la columna `year_of_release`, aprovecharemos la relación entre el año de lanzamiento y la plataforma del juego. Nuestro enfoque implicará utilizar la información de la plataforma de la columna `platform` como referencias parciales.

Para lograr esto, definiremos una función personalizada que tome la plataforma como entrada y devuelva el valor medio de `year_of_release` para esa plataforma específica. Esta función personalizada nos permitirá completar los valores faltantes en función del año de lanzamiento medio asociado con cada plataforma.

Al aplicar esta función personalizada a los valores faltantes en la columna `year_of_release`, podemos completar eficazmente los espacios en blanco en función del año de lanzamiento medio para cada plataforma correspondiente.

```python
# Definir una función para completar los valores faltantes

def fill_missing_value(data, agg_column, value_column):
    grouped_values = data.groupby(agg_column)[value_column].median().reset_index()
    size = len(grouped_values)
    for i in range(size):
        group = grouped_values[agg_column][i]
        value = grouped_values[value_column][i]
        data.loc[(data[agg_column]==group) & (data[value_column].isna()), value_column] = value
    return data
```

```python
# Implementar la función en las columnas 'platform' y 'year_of_release'

fill_missing_value(df, 'platform', 'year_of_release')
```

```python
# Verificando el resultado

df['year_of_release'].isna().sum()
```

Ahora cambiamos el tipo de datos

```python
# Cambiando el tipo de datos

df['year_of_release'] = df['year_of_release'].astype('int')
```

##### Columna `rating`


Primero comprobaremos los valores únicos incluidos en esta columna.

```python
# Comprobando si hay valores únicos en la columna `rating`

df['rating'].unique()
```

La columna `rating` no contiene datos numéricos, por lo que no podemos utilizar la mediana para completar los valores faltantes. En su lugar, crearemos una nueva función que utilice la moda (el valor que aparece con mayor frecuencia) y la asociaremos con la columna `genre`.

```python
# Creamos una nueva función para completar los valores faltantes usando el modo

def fill_missing_rating(data, agg_column, value_column):
    grouped_values = data.groupby(agg_column)[value_column].agg(pd.Series.mode).reset_index()
    size = len(grouped_values)
    for i in range(size):
        group = grouped_values[agg_column][i]
        value = grouped_values[value_column][i]
        data.loc[(data[agg_column]==group) & (data[value_column].isna()), value_column] = value
    return data
```

```python
# Implementar la función en la columna de género y calificación

fill_missing_rating(df, 'genre', 'rating')
```

```python
# Verificando el resultado

df['rating'].isna().sum()
```

Investigando un poco encontramos que la clasificación "K-A" o "Niños a adultos" es otro nombre para "E" (Todos). La clasificación "K-A" se utilizó antes de 1999 hasta que finalmente se cambió a "E". Por lo tanto, reemplazaremos "K-A" por "E" en la columna de "clasificación".

```python
# Cambiar el valor en la columna `rating`

df['rating'] = df['rating'].replace({'K-A' : 'E'})
df['rating'].unique()
```

## Prueba de hipótesis

Hemos completado el proceso de análisis de datos, hemos obtenido hallazgos reveladores y brindado recomendaciones para el marketing de videojuegos en 2017. Ahora, es momento de realizar pruebas de hipótesis utilizando métodos estadísticos.

Se nos han asignado dos hipótesis para probar:

1. Hipótesis nula (H0): la calificación promedio de los usuarios para las plataformas Xbox One y PC es la misma.
Hipótesis alternativa (H1): la calificación promedio de los usuarios para las plataformas Xbox One y PC es diferente.

2. Hipótesis nula (H0): la calificación promedio de los usuarios para los géneros Acción y Deportes es la misma.
Hipótesis alternativa (H1): la calificación promedio de los usuarios para los géneros Acción y Deportes es diferente.

### Hipótesis 1: La calificación promedio de los usuarios para las plataformas Xbox One y PC es la misma.

La hipótesis nula (H0) y la hipótesis alternativa (H1) para la prueba son las siguientes:
- H0: La calificación promedio de los usuarios para las plataformas Xbox One y PC es la misma.
- H1: La calificación promedio de los usuarios para las plataformas Xbox One y PC es diferente.

Para probar estas hipótesis, utilizaremos el método `st.ttest_ind()`, que compara las medias de dos muestras independientes. Primero, crearemos poblaciones de datos separadas para las plataformas Xbox One y PC en función de la columna `user_rating`. Utilizaremos los datos de la población para todo el año para asegurar un tamaño de muestra adecuado para la prueba.

Antes de realizar la prueba de hipótesis, debemos verificar si las varianzas de las dos poblaciones son iguales. Esto se puede hacer utilizando una prueba estadística adecuada.

El valor alfa (nivel de significancia) para la prueba se establece en 5%. Si el valor p resultante de la prueba es menor que 0.05, rechazaremos la hipótesis nula (H0) y concluiremos que las calificaciones promedio de los usuarios para las dos plataformas son diferentes. Por el contrario, si el valor p es mayor o igual a 0.05, aceptaremos la hipótesis nula (H0) y concluiremos que las calificaciones promedio de los usuarios para las dos plataformas son las mismas.

```python
# Crear población de datos para la plataforma Xbox One

df_xone = df.query('platform == "XOne"')
df_xone
```

```python
# Crear población de datos para plataformas de PC

df_pc = df.query('platform == "PC"')
df_pc
```

**A continuación, examinaremos los valores de varianza de estos dos conjuntos de datos**

```python
df_xone['user_score'].var()
```

```python
df_pc['user_score'].var()
```

Dado que los valores de varianza de los dos conjuntos de datos no son significativamente diferentes, asumiremos que son iguales en la prueba de hipótesis. Por lo tanto, estableceremos el parámetro 'equal_var' como verdadero.

Ahora, procedamos con la prueba de hipótesis.

```python
alpha = 0.05
results = st.ttest_ind(df_xone['user_score'], df_pc['user_score'], equal_var=True)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print("Rechazamos H0")
else:
    print("Aceptamos H0")
```

Con base en los resultados de la prueba de hipótesis, rechazamos la hipótesis nula (H0) y aceptamos la hipótesis alternativa (H1).

Por lo tanto, podemos concluir que **la calificación promedio de los usuarios para las plataformas Xbox One y PC es diferente.**

### Hipótesis 2: Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son diferentes. <a id="Hipótesis-2-Las-calificaciones-promedio-de-los-usuarios-para-los-géneros-Accion-y-Deportes-son-diferentes"></a>


La siguiente es la hipótesis nula (H0) y la hipótesis alternativa (H1) que probaremos:

- H0: La calificación promedio de los usuarios para los géneros Acción y Deportes es la misma.
- H1: La calificación promedio de los usuarios para los géneros Acción y Deportes es diferente.

Crearemos un nuevo conjunto de datos que incluya solo los géneros "Acción" y "Deportes". Luego, realizaremos una prueba de hipótesis basada en la columna `user_score` para probar si la calificación promedio de los usuarios difiere entre estos dos géneros.

```python
# Crear población de datos para el género Acción

df_action = df.query('genre == "Action"')
df_action
```

```python
# Crear población de datos para el género Deportes

df_sports = df.query('genre == "Sports"')
df_sports
```

**A continuación, examinaremos los valores de varianza de estos dos conjuntos de datos**

```python
df_action['user_score'].var()
```

```python
df_sports['user_score'].var()
```

Los valores de varianza de los dos conjuntos de datos tampoco son significativamente diferentes, por lo que procederemos con la prueba de hipótesis asumiendo varianzas iguales.

Ahora, realizaremos la prueba de hipótesis para determinar si la calificación promedio de los usuarios para los géneros Acción y Deportes es diferente.

```python
alpha = 0.05
results = st.ttest_ind(df_action['user_score'], df_sports['user_score'], equal_var=True)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print("Rechazamos H0")
else:
    print("Aceptamos H0")
```

Con base en los resultados de la prueba de hipótesis, hemos decidido aceptar la hipótesis nula (H0). **Por lo tanto, concluimos que la calificación promedio de los usuarios para los géneros Acción y Deportes es la misma.**
